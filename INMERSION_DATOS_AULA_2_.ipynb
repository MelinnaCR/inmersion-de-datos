```python
inmuebles.sample(10)
inmuebles.Tipo
inmuebles_tres_tipos = inmuebles[inmuebles.Tipo.isin(['Apartamento','Casa','Bodega'])]
inmuebles_tres_tipos
```

```python
from matplotlib import style
plt.figure(figsize=(10,5))
style_list = ['defaul', 'dark_background'] + sorted(style for style in plt.style.available if style != 'dark_background')
plt.style.use(style_list[1])
grafica_tres_tipos= sns.histplot(data=inmuebles_tres_tipos, x='Precio_Millon', kde=True, hue='Tipo')
grafica_tres_tipos.set_title('Distribucion de Valor de los inmuebles de Apartamento, Casa, Bodega', color='red', fontsize=15)
plt.xlim((50,1000))
plt.ylim(0,750)
grafica.set_xlabel('Precio_Millon', fontsize=18)
grafica.set_ylabel('Count', fontsize=16)
plt.savefig('/content/valor_data.png',format = 'png')
plt.show()

```

**CONCLUSIÓN**: En el gráfico anterior se puede apreciar como la distribución de valor de los inmuebles de Bogotá, se divide entre apartamentos y casas con mas de 2000 pesos colombianos. Por otro lado las Bodegas es inferior a dichos inmobiliarios con un valor de 150 pesos colombianos.

```python
 
from pickle import NONE
inmuebles_barrio_m2 = inmuebles.groupby('Barrio').mean().head(15)
inmuebles_barrio_m2['Precio_m2'] = inmuebles_barrio_m2['Precio_Millon'] / inmuebles_barrio_m2['Area']
inmuebles_barrio_m2.sort_values(by='Precio_m2', ascending=False, inplace=True)

plt.figure(figsize=(10 ,6))
grafica = sns.barplot(data=inmuebles_barrio_m2, y=inmuebles_barrio_m2.index, x='Precio_m2', palette='coolwarm',orient='h' )

grafica.set_title('DISTRIBUCIÓN DE PRECIO POR METRO CUADRADO', color='purple', fontsize=20)
grafica.set_xlabel('Precio en millones')
grafica.set_ylabel('Barrio')

plt.show()
```
